# Data Preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('AirPassengers.csv') #read the dataset
```

```python
dataset.head()
```

```python
dataset.plot(figsize=(12,4))
```

```python
dataset.isnull().any()
```

```python
#splitting the year,month and day into seperate columns
dataset['Year'] = pd.DatetimeIndex(dataset['Month']).year 
dataset['Mon'] = pd.DatetimeIndex(dataset['Month']).month
dataset['Day'] = pd.DatetimeIndex(dataset['Month']).day
```

```python
dataset.head()
```

```python
dataset.drop('Month',axis=1,inplace=True) # deleting the date column as it is unnecessary
```

```python
dataset.head()
```

```python
dataset.isnull().any() 
```

```python
dataset.isnull().sum()
```

```python
#handling of null values
dataset['#Passengers'].fillna(dataset['#Passengers'].mean(),inplace=True) 
```

```python
dataset.isnull().any()
```

# Data visualization

```python
import seaborn as sns
```

```python
x=dataset.iloc[:,1:4].values
y=dataset.iloc[:,0:1].values
```

```python
x
```

```python
y
```

```python
#splitting the data into train and test dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#standard scaling for better model training
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Creating the model for price prediction-regression

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense, Dropout
```

```python
reg=Sequential()
```

```python
reg.add(Dense(units=3,kernel_initializer="random_uniform",activation="relu"))
```

```python
reg.add(Dense(units=100,kernel_initializer="random_uniform",activation="relu"))
```

```python
reg.add(Dense(units=100,kernel_initializer="random_uniform",activation="relu"))
```

```python
reg.add(Dense(units=1,kernel_initializer="random_uniform"))
```

```python
reg.compile(optimizer="rmsprop",loss="mse",metrics=['mse'])
```

```python
history= reg.fit(x_train,y_train,batch_size=200,epochs=4000,validation_data=(x_test,y_test))
```

```python
y_pred=reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
accuracy=r2_score(y_test,y_pred)
accuracy
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'],label="train_loss")
plt.plot(history.history['val_loss'],label='test_loss')
plt.legend()
plt.show()
```
